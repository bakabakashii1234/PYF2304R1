### Bài 5: Viết chương trình nhập vào một số bất kỳ và in ra giá trị tuyệt đối của số đó

giá trị tuyệt đối của một số là khoảng cách từ số đó đến 0



````python
number = int(input("Nhập vào 1 số: "))
if number >= 0:
    print(f"giá tri tuyệt đối của {number}: {number-0}")
else:
    print(f"giá tri tuyệt đối của {number}: {0-number}")
````

### Bài 6: Học viên sẽ không đủ điều kiện thi nếu tỉ lệ có mặt nhỏ hơn 75%. Viết chương trình nhập vào tổng số buổi học của một khóa học và tổng số buổi học mà học viên tham gia. Sau đó cho biết học viên đó có đủ điều kiện thi hay không.

````python
sum_lesson = int(input("Enter total lesson: "))
current_sum_lesson = int(input("Enter current total lesson: "))
percent = int(current_sum_lesson/sum_lesson*100)
if percent < 75:
    print(f"rate: {percent}% ineligible for the exam")
else:
    print(f"rate: {percent}% qualified for the exam")
````

### Bài 7: Một công ty vận tải có cách tính chi phí giao hàng như sau:
1. 10kilogram đầu tiên: \$5/kg
2. Từ kilogram thứ 11 đến 20: \$7/kg
3. Từ kilogram thứ 20 trở lên: \$10/kg

Viết chương trình nhập vào số kg cần giao(số nguyên) và in ra chi phí giao hàng tương ứng

````python
kg = int(input("Enter kilogram want to send: "))
if kg <= 10:
    print("condition for the first 10kilogram: $5/kg")
    print(f"current kilogram: {kg}, cost: {kg*5}$")
elif 10 < kg < 20:
    print("condition for 11kilogram -> 20kilogram: $7/kg")
    print(f"current kilogram: {kg}, cost: {kg*7}$")
else:
    print("condition for 20kilogram more: $10/kg")
    print(f"current kilogram: {kg}, cost: {kg*10}$")
````

### Bài 8: Viết chương trình nhập vào số năm và cho biết năm đó là năm nhuận hay năm không nhuận.

````python
year = int(input("Enter the year(yyyy) you want: "))
if year%4==0:
    print(f"{year} year is leap year")
else:
    print(f"{year} year is not leap year")
````

### Bài 10:
Tại một cửa hàng nọ có menu như sau:

1. Gà rán giá 35k
2. Hamburger giá 50k
3. Pesi giá 10k

Nhập số lượng món hàng cần mua

Tính và in ra tổng giá tiền

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"

````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""
````

Menu cửa hàng:
 1. Gà rán giá 35k
 2. Hamburger giá 50k
 3. Pesi giá 10k

````python
ga_ran = int(input("Hãy nhập số lượng gà rán muốn mua: "))
hambuger = int(input("Hãy nhập số lượng hambuger muốn mua: "))
pesi = int(input("Hãy nhập số lượng pesi muốn mua: "))
tong_gia = ga_ran*35000 + hambuger*50000 + pesi*10000
print(f"tổng giá tiền đơn hàng: {tong_gia: ,} đồng")
if 2000 <tong_gia <= 4000000:
    print(f"tổng tiền được giảm: {tong_gia*0.1:,.0f} đồng")
    print(f"tổng tiền phải thanh toán: {tong_gia-tong_gia*0.1:,.0f} đồng")
elif tong_gia > 4000000:
    print(f"tổng tiền được giảm: {tong_gia*0.15:,.0f} đồng")
    print(f"tổng tiền phải thanh toán: {tong_gia-tong_gia*0.15:,.0f} đồng")
else:
    print(f"Tổng tiền phải thanh toán là: {tong_gia} đồng")
print("Cảm ơn quý khách đã mua")
````
