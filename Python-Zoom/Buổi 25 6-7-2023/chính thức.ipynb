```python
class Ingredients:
    def __init__(self, name, link):
        self.name = name
        self.link = link
```

```python
from tkinter import *
from tkinter import messagebox
import webbrowser as wb
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo
import pickle
from PIL import ImageTk, Image 
import tkinter

a = "skyblue"
#tạo cửa sổ
window = Tk()
#tạo size cửa sổ
window.minsize(height = 600, width = 800 )
style = ttk.Style(window)
style.theme_use("clam")
style.configure("Treeview", background=a, 
                fieldbackground= a, foreground="black")
#tạo Treeview
window.title("ỨNG DỤNG QUẢN LÍ BÀI HÁT")
window.configure(background = a)
tree = ttk.Treeview(window, columns=("name","link"), show='headings')

# define headings
tree.heading("name", text='Tên bài hát')
tree.heading("link", text='Đường dẫn')

ingredient_list = []
def add_ingredient():
    name_value= name.get()
    link_value= link.get()
    if name_value== "" or link_value== "":
        messagebox.showerror("Bài hát", "Xin vui lòng nhập bài hát và đường dẫn!")
    else:
        #lấy giá trị nhập vào đưa vào class
        new_ingredient= Ingredients(name_value,link_value)
        #Lưu vào list để hiển thị
        ingredient_list.append(new_ingredient)
        insert_to_treeview()
        messagebox.showinfo("Bài hát", f"Thêm bài hát: {name_value} thành công!")
def delete_ingredient():
    try:
        #chọn item trong tree
        selected = tree.selection()[0]
        #dạng tuple
        data = (selected,)
        #chọn item có địa chỉ data, lấy giá trị (values) ở cột 0
        delete_name = tree.item(data)["values"][0]
        for i in range(len(ingredient_list)):
            if ingredient_list[i].name == delete_name:
                del ingredient_list[i]
                tree.delete(selected)
                messagebox.showinfo("Xóa bài hát", f"Xóa bài hát {delete_name} thành công!")
                insert_to_treeview()
                break
    except:
        messagebox.showerror("Xóa bài hát", "Xin hãy chọn bài hát!")     
def search_ingredient():
    try:    
        search_name = search.get()
        check = False
        clear_tree() 
        for i in range(len(ingredient_list)):
            if ingredient_list[i].name == search_name:
                tree.insert('', 'end', values=(ingredient_list[i].name, ingredient_list[i].link))
                check = True
        if check == False:
            messagebox.showeinfo("Tìm kiếm bài hát", "Không tìm thấy bài hát!")
            insert_to_treeview()
    except:
        messagebox.showerror("Tìm kiếm bài hát", "Xin hãy nhập bài hát!")
        insert_to_treeview()
def update_link():
    try:
        update_link = new_link.get()
        #lấy item ở treeview
        selected = tree.selection()[0]
        data = (selected,)
        search_name = tree.item(data)["values"][0]
        if update_link== "":
            messagebox.showerror("Bài hát", "Xin vui lòng nhập đường dẫn")
        else:
            for i in range(len(ingredient_list)):
                if ingredient_list[i].name == search_name:
                    ingredient_list[i].link = update_link
                    messagebox.showinfo("Cập nhật đường dẫn",f"Tên bài hát: {ingredient_list[i].name} \n" 
                                                             f"Đường dẫn đã được cập nhật: {ingredient_list[i].link}")
                    insert_to_treeview()
                    break
    except:
        messagebox.showerror("Xóa bài hát", "Xin hãy chọn bài hát!")   
def sort_ingredient():
    check = False
    for i in range(len(ingredient_list)-1):
        for j in range(i + 1, len(ingredient_list)):
            if ingredient_list[i].name > ingredient_list[j].name:
                c = ingredient_list[i]
                ingredient_list[i] = ingredient_list[j]
                ingredient_list[j] = c
                check = True
    if check == True:
        messagebox.showinfo("Sắp xếp bài hát", "Sắp xếp bài hát thành công!")
        insert_to_treeview()
#đưa data vào file
def save_tree():
    f = open("video.txt", "w", encoding = "utf8")
    for i in range(len(ingredient_list)-2):
        f.write(ingredient_list[i].name)
        print(ingredient_list[i].name)
        f.write("\n")
        f.write(ingredient_list[i].link)
        print(ingredient_list[i].link)
        f.write("\n")
    #Để tránh trường hợp xuống dòng ở phần tử cuối list
    f.write(ingredient_list[len(ingredient_list)-1].name)
    f.write("\n")
    f.write(ingredient_list[len(ingredient_list)-1].link)
    f.close()
#lấy data từ file
try:
    f = open("video.txt", "r", encoding = "utf8") 
    list1 = f.read().split("\n")
    for i in range(0,len(list1),2):
        name_video = list1[i]
        link_video = list1[i+1]
        new_video = Ingredients(name_video, link_video)
        ingredient_list.append(new_video)
    f.close()
    insert_to_treeview()
except:
    messagebox.showerror("Danh sách bài hát", "Không tải được danh sách bài hát!")
def clear_tree():
    for i in tree.get_children():
        tree.delete(i)
def insert_to_treeview():
    clear_tree()
    for i in range(len(ingredient_list)):
        tree.insert('', 'end', values=(ingredient_list[i].name, ingredient_list[i].link))
def back_to_tree():
    insert_to_treeview() 
def open_link(event):
    tree = event.widget  # get the treeview widget
    #link chú thích hàm: https://w10schools.com/posts/167348_tkinter%252Ettk%252ETreeview%252Eidentify_region%28%29
    region = tree.identify_region(event.x, event.y)
    col = tree.identify_column(event.x)
    iid = tree.identify('item', event.x, event.y)
    if col == '#2':
        link = tree.item(iid)['values'][1]  # get the link from the selected row
        wb.open_new_tab(link)  # open the link in a browser tab
# bind the selection event to 'open_link'  
#button-1 dùng để ấn chuột
tree.bind('<Button-1>', open_link)        
# add a scrollbar
scrollbar = ttk.Scrollbar(window, orient=tk.VERTICAL)
tree.configure(yscrollcommand = scrollbar.set)
scrollbar.config(command = tree.yview)
# Create a photoimage object of the image in the path
image1 = Image.open(r"E:\PYTHON-codegym\PYF2304R1\Python-Zoom\Buổi 25 6-7-2023\image.JPG")
test = ImageTk.PhotoImage(image1)
label1 = tkinter.Label(image=test)
label1.image = test
#Tạo label
lbl_title = Label(window, text = "BALLAD-nhẹ nhàng sâu lắng", fg="red", bg= a, font = ("cambria",17))
lbl_title1 = Label(window, text = "Playlist", fg = "black",bg = a, font = ("cambria",11))
lbl_name = Label(window, text = "Tên bài hát: ", fg = "black",bg = a)
lbl_link = Label(window, text = "Đường dẫn: ", fg = "black",bg = a)
lbl_name_search = Label(window, text = "Tên bài hát: ", fg = "black",bg =a)
lbl_new_link = Label(window, text = "Đường dẫn mới: ", fg = "black", bg = a)


#Tạo entry
name = Entry(window)
link = Entry(window)
search = Entry(window)
new_link = Entry(window)
#Tạo button
btn_add = Button(window, text = "Thêm", height = 1, width = 15, activebackground = "blue", command = add_ingredient)
btn_delete = Button(window, text = "Xóa", height = 1, width = 15, activebackground = "blue", command = delete_ingredient)
btn_search = Button(window, text = "Tìm kiếm",height = 1, width = 15, activebackground = "blue", command = search_ingredient)
btn_sort = Button(window, text = "Sắp xếp",height = 1, width = 15, activebackground = "blue", command = sort_ingredient)
btn_update = Button(window, text = "Cập nhật đường dẫn",height = 1, width = 17, activebackground = "blue", command = update_link)
btn_back_tree = Button(window, text = "Trở lại danh sách",height = 1, width = 17, activebackground = "blue", command = back_to_tree)
btn_save_tree = Button(window, text = "Lưu danh sách",height = 1, width = 17, activebackground = "blue", command = save_tree)
btn_quit = Button(window, text = "Thoát",height = 1, width = 17, activebackground = "blue", command = window.destroy)
#Tạo nơi để đặt vào

btn_add.place(x = 60, y = 40)
lbl_name.place(x = 10, y = 80)
name.place(x = 90, y = 80)
lbl_link.place(x= 10, y = 120)
link.place(x = 90, y = 120)

btn_search.place(x = 20, y = 180)
btn_back_tree.place(x = 150, y = 180)
lbl_name_search.place(x = 10, y = 220)
search.place(x = 90, y= 220)


btn_update.place(x = 60, y = 280)
lbl_new_link.place(x = 5, y = 320)
new_link.place(x = 110, y= 320)

btn_delete.place(x = 60, y = 360)
btn_sort.place(x = 60, y = 400)
btn_save_tree.place(x = 60, y = 440)
btn_quit.place(x=60, y= 480)
label1.place(x=350, y = 40)
lbl_title1.place(x=460, y= 80)
lbl_title.place(x = 460, y= 100)
tree.place(x = 350, y = 150)
scrollbar.place(x = 755, y=151, height = 223)
#kích hoạt window
window.mainloop()
```
